```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter
```

```python
y = np.genfromtxt('seasonal_data/irradiance.txt')

# y = savgol_filter(y, window_length=169, polyorder=2)

x = np.arange(0, len(y))

horas_por_dia = 24
num_dias = len(y) // horas_por_dia
y_por_dia = y[:num_dias * horas_por_dia].reshape(num_dias, horas_por_dia)

dias = np.arange(1, num_dias + 1)

media_diaria = y_por_dia.mean(axis=1)

# Cria o gráfico
plt.figure(figsize=(8, 4))
plt.plot(dias, media_diaria, color='tomato', linewidth=1,
         markerfacecolor='white', markeredgecolor='tomato')

# Títulos e legendas
plt.xlabel('Dia', fontdict={'size': 14})
plt.ylabel('Irradiância (kW/m^2)', fontdict={'size': 14})
plt.yticks(fontsize=12)
plt.xticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.gca().set_facecolor('#f9f9f9')
plt.box(False)
plt.tick_params(axis='both', labelsize=10)
plt.tight_layout()
```
